#### 12. [0.75 баллов] Предположим, что в данных была ошибка и Diet Coke (choice_description), который стоил $1.25, должен был стоить 1.35. Скорректируйте данные в таблицы и посчитайте, на какой процент больше денег было заработано с этого товара. Не забывайте, что количество товара не всегда равно 1.

```python
# Находим Diet Coke que cuestan exactamente $1.25 por unidad

diet_coke_mask = (pass_data['choice_description'].str.contains('Diet Coke', na=False)) & (pass_data['unit_price'] == 1.25)
diet_coke_data = pass_data[diet_coke_mask]

if len(diet_coke_data) == 0:
    print("No se encontraron Diet Coke con precio de $1.25")
else:
    # Считаем текущую выручку de los Diet Coke de $1.25
    current_revenue = diet_coke_data['price_numeric'].sum()
    total_units = diet_coke_data['quantity'].sum()
    
    print("АНАЛИЗ DIET COKE DE $1.25:")
    print(f"Количество записей Diet Coke по $1.25: {len(diet_coke_data)}")
    print(f"Общее количество проданных единиц: {total_units}")
    print(f"Текущая выручка: ${current_revenue:.2f}")
    
    # Actualizar price_numeric
    pass_data.loc[diet_coke_mask, 'price_numeric'] = 1.35 * pass_data.loc[diet_coke_mask, 'quantity']
    
    # Si existe unit_price, actualizarla también
    pass_data.loc[diet_coke_mask, 'unit_price'] = 1.35
    
    # Новая выручка de los mismos items
    new_revenue = pass_data.loc[diet_coke_mask, 'price_numeric'].sum()
    
    # Расчет увеличения
    revenue_increase = new_revenue - current_revenue
    percentage_increase = (revenue_increase / current_revenue) * 100
    
    print("\nПОСЛЕ КОРРЕКТИРОВКИ:")
    print(f"Новая выручка: ${new_revenue:.2f}")
    print(f"Увеличение выручки: ${revenue_increase:.2f}")
    print(f"Процент увеличения: {percentage_increase:.2f}%")
    
```

## 2. Визуализации и matplotlib

При работе с данными часто неудобно делать какие-то выводы, если смотреть на таблицу и числа в частности, поэтому важно уметь визуализировать данные. В этом разделе мы этим и займёмся.

У matplotlib, конечно, же есть [документация](https://matplotlib.org/users/index.html) с большим количеством [примеров](https://matplotlib.org/examples/), но для начала достаточно знать про несколько основных типов графиков:
- plot — обычный поточечный график, которым можно изображать кривые или отдельные точки;
- hist — гистограмма, показывающая распределение некоторое величины;
- scatter — график, показывающий взаимосвязь двух величин;
- bar — столбцовый график, показывающий взаимосвязь количественной величины от категориальной.

В этом задании вы попробуете построить каждый из них. Не менее важно усвоить базовые принципы визуализаций:
- на графиках должны быть подписаны оси;
- у визуализации должно быть название;
- если изображено несколько графиков, то необходима поясняющая легенда;
- все линии на графиках должны быть чётко видны (нет похожих цветов или цветов, сливающихся с фоном);
- если отображена величина, имеющая очевидный диапазон значений (например, проценты могут быть от 0 до 100), то желательно масштабировать ось на весь диапазон значений (исключением является случай, когда вам необходимо показать малое отличие, которое незаметно в таких масштабах).
- сетка на графике помогает оценить значения в точках на глаз, это обычно полезно, поэтому лучше ее отрисовывать.

#### 3. [1 балл] Постройте средних сумм покупок по дням недели (bar plot).

```python
order_totals = pass_data.groupby('order_id').agg({
    'price_numeric': 'sum',
    'date': 'first'  # Tomamos la primera fecha del pedido
}).reset_index()

order_totals.head()
```

```python
# Convertir la columna 'date' a datetime
order_totals['date'] = pd.to_datetime(order_totals['date'])

# Añadimos el día de la semana a cada pedido
order_totals['day_of_week'] = order_totals['date'].dt.day_name()
order_totals.head()
```

```python
# Calculamos el promedio de las sumas de pedidos por día de la semana
weekday_avg_totals = order_totals.groupby('day_of_week')['price_numeric'].mean()

# Упорядочиваем дни недели
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_avg_totals = weekday_avg_totals.reindex(weekday_order)
weekday_avg_totals
```

```python
# График
plt.figure(figsize=(8, 4))
weekday_avg_totals.plot(kind='bar', color='lightgreen')
plt.xlabel('День недели')
plt.ylabel('Средняя сумма заказа')
plt.title('Средние суммы покупок по дням недели')

min_value = weekday_avg_totals.min()
max_value = weekday_avg_totals.max()
plt.ylim(17, max_value + 0.5)  # Empezar en 17
plt.yticks(np.arange(17, max_value + 0.5, 0.5))  # Incrementos de 0.5
plt.xticks(rotation=0)

# Agregar los valores encima de cada barra
for i, (day, value) in enumerate(weekday_avg_totals.items()):
    plt.text(i, value + 0.1, f'${value:.2f}', ha='center', fontsize=10)

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```
